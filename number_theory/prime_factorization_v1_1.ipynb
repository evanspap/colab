```python
print("="*60)
print(" PRIME FACTORIZATION ‚Äî STEP BY STEP (Educational)")
print("="*60)
print("This notebook shows how an integer is factorized")
print("into prime numbers, step by step.\n")
```

```python
n = int(input("ŒîœéœÉŒµ Œ≠ŒΩŒ±ŒΩ Œ∏ŒµœÑŒπŒ∫œå Œ±Œ∫Œ≠œÅŒ±ŒπŒø (>1): "))

if n <= 1:
    raise ValueError("Œü Œ±œÅŒπŒ∏ŒºœåœÇ œÄœÅŒ≠œÄŒµŒπ ŒΩŒ± ŒµŒØŒΩŒ±Œπ ŒºŒµŒ≥Œ±ŒªœçœÑŒµœÅŒøœÇ œÑŒøœÖ 1.")

original_n = n

print(f"\nŒëœÅœáŒπŒ∫œåœÇ Œ±œÅŒπŒ∏ŒºœåœÇ: {n}")
print("-"*60)
```

```python
factors = {}          # {prime: exponent}
tested_primes = []    # all divisors tested
found_primes = []     # primes that actually divide n
```

```python
print("‚û° ŒàŒªŒµŒ≥œáŒøœÇ Œ≥ŒπŒ± œÑŒøŒΩ œÄœÅœéœÑŒø 2")

while n % 2 == 0:
    print(f"  {n} √∑ 2 = {n//2}")
    factors[2] = factors.get(2, 0) + 1
    n //= 2
    found_primes.append(2)

tested_primes.append(2)

if 2 in factors:
    print(f"‚úî ŒíœÅŒ≠Œ∏Œ∑Œ∫Œµ: 2^{factors[2]}")
else:
    print("‚úò Œ§Œø 2 Œ¥ŒµŒΩ Œ¥ŒπŒ±ŒπœÅŒµŒØ œÑŒøŒΩ Œ±œÅŒπŒ∏Œºœå")

print(f"Œ•œÄœåŒªŒøŒπœÄŒø n = {n}")
print("-"*60)
```

```python
# We start with p = 3 because:
# - Factor 2 has already been handled earlier
# - All even numbers > 2 are composite
p = 3

# Informational output:
# We are about to test odd candidate divisors up to sqrt(n)
print("‚û° Checking odd candidate divisors up to ‚àön\n")

# Main trial division loop
# ------------------------------------------------
# Condition: p * p <= n
#
# Mathematical meaning:
# If n has a composite factorization n = a * b,
# then at least one of a or b must be <= sqrt(n).
#
# Therefore:
# - If no divisor p <= sqrt(n) is found,
# - then the remaining n must be prime.
#
while p * p <= n:

    # Record that this candidate p was TESTED
    # This happens for EVERY p that enters the loop,
    # regardless of whether it divides n or not.
    tested_primes.append(p)

    # Check whether p actually divides n
    if n % p == 0:
        print(f"‚úî {p} divides n")

        # Inner loop:
        # As long as p continues to divide n,
        # we keep removing p from n.
        #
        # This counts the exponent of p in the prime factorization.
        while n % p == 0:
            # Print the division that is ABOUT to happen
            # (this line does not modify n)
            print(f"  {n} √∑ {p} = {n//p}")

            # Increase the exponent of prime factor p
            #
            # factors.get(p, 0):
            #   - returns current exponent if p exists
            #   - returns 0 if p is not yet in the dictionary
            #
            # Then we add 1 to account for this division
            factors[p] = factors.get(p, 0) + 1

            # Actually remove one factor of p from n
            n //= p

            # Record that p was FOUND as a prime factor
            # (this list may contain duplicates)
            found_primes.append(p)

        # At this point, p no longer divides n
        # We have fully extracted p^k from the original number
        print(f"  ‚Üí Total power: {p}^{factors[p]}")
        print(f"  ‚Üí Reduced n = {n}\n")

    else:
        # p was tested but does not divide n
        print(f"‚úò {p} does not divide n")

    # Move to the next odd candidate divisor
    # (skip even numbers entirely)
    p += 2

# Visual separator after finishing the trial-division phase
print("-" * 60)
```

```python
if n > 1:
    print(f"‚û° Œ§Œø œÖœÄœåŒªŒøŒπœÄŒø {n} ŒµŒØŒΩŒ±Œπ œÄœÅœéœÑŒøœÇ")
    factors[n] = factors.get(n, 0) + 1
    found_primes.append(n)
else:
    print("‚û° ŒîŒµŒΩ Œ≠ŒºŒµŒπŒΩŒµ œÖœÄœåŒªŒøŒπœÄŒøœÇ œÄœÅœéœÑŒøœÇ")

print("-"*60)
```
