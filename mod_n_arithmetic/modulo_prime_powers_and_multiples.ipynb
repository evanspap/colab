# Υπολογισμοί modulo p

Αυτό το notebook:
- Παίρνει έναν **πρώτο αριθμό p**
- Παίρνει έναν ακέραιο **α με 1 ≤ α ≤ p−1**
- Υπολογίζει:
  1. Όλες τις δυνάμεις **α^k (mod p)** για k = 1 έως p−1
  2. Όλα τα πολλαπλάσια **k·α (mod p)** για k = 1 έως p−1

Κατάλληλο για χρήση στο **Google Colab**.


```python

# Είσοδοι
p = 23   # πρώτος αριθμός
a = 4    # ακέραιος με 1 <= a <= p-1

assert p > 2, "Το p πρέπει να είναι πρώτος > 2"
assert 1 <= a <= p-1, "Το a πρέπει να είναι στο [1, p-1]"
```

```python

print("Δυνάμεις του a modulo p")
print("----------------------")
for k in range(1, p):
    value = pow(a, k, p)
    print(f"a^{k} mod {p} = {value}")
```

```python

print("\nΠολλαπλάσια του a modulo p")
print("--------------------------")
for k in range(1, p):
    value = (k * a) % p
    print(f"{k}·a mod {p} = {value}")
```
