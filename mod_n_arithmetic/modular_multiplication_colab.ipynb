```python
# Define modulus n
n = 180  # change this to any positive integer
```

```python
# Take inputs
a = int(input('Enter a: '))
```

```python
b = int(input('Enter b: '))
```

```python
# Compute modular addition
result = (a * b) % n
print(f"({a} * {b}) mod {n} = {result}")
```
